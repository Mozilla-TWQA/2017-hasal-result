```python
import pandas as pd
import numpy as np
import json 
import matplotlib.pyplot as plt
from os import environ


pd.options.display.float_format = '{:10,.2f}'.format
pd.set_option('display.width', 999)

key_index = ['actual_paint_frames',
            'expected_frames', 
            'frame_throughput',
            'long_frame']

with open('/Users/hasal/PycharmProjects/2017-hasal-result/20170510/facebook/win7/report-FT-facebook-composer-input-100-chars-type-interval-50-with-threshold.json') as data_file:
    data = json.load(data_file)
    d = pd.DataFrame(data)
    df = pd.concat([pd.DataFrame(d[column]['time_list']).T for column in d], axis=1)
    df.columns = d.columns

    dft = df.loc[key_index, :]
    
dft
```

```python
frame = pd.DataFrame([pd.Series(df[c]['time_sequence']).diff() \
                                                       .apply(lambda x: np.round(x/11.11)) \
                                                       .drop(0) \
                                                       .astype(int) for c in df]).T
frame.columns = df.columns
frame
```
